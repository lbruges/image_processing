```python
#Ejemplo practico con Python 

#Se utiliza matplotlib para visualizar la imagen
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray

#imagen original de data
original = data.astronaut()
#Imagen en escala de grises
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
#Ejemplo:
#Visualizando en el histograma la cantidad de rojo que tiene la imagen
red = original[:, :, 0]
# Obtain the red histogram
plt.hist(red.ravel(), bins=256, color='red')
plt.show()
```

```python
#Visualizando la cantidad de Verde
green = original[:, :, 1]
plt.hist(green.ravel(), bins=256, color='green')
plt.title('Green Histogram')
plt.show()
```

```python
#Visualizando la cantidad de azul 
blue = original[:, :, 2]
plt.hist(blue.ravel(), bins=256)
plt.title('Blue Histogram')
plt.show()
```

# Umbrales

Particionar una imagen en un primer plano. Haciéndolo en blanco y negro configurando cada píxel en:
255 (blanco) si píxel> valor de umbral
0 (negro) si píxel < valor de umbral



```python
from skimage.filters import try_all_threshold
# Obtain all the resulting images
fig, ax = try_all_threshold(grayscale, verbose=False)
# Showing resulting plots
plt.show(fig, ax)
```

# Filtros

Mejorar una imagen
Enfatizar o eliminar características
Suavizado
Afilado
Detección de bordes

```python
# comparacion 
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
    sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered)
    ax2.axis('off')
```

```python
# Import the module and function
#####Aplicando un gaussiano#####
from skimage.filters import gaussian
# Apply edge detection filter
gaussian_image = gaussian(original, multichannel=True)
# Show original and resulting image to compare
plot_comparison(original, gaussian_image, "Blurred with Gaussian filter")
```

# Transformando imagenes

Optimización y compresión de imágenes
Guarda imágenes con la misma proporción
Preparando imágenes para clasificacion de modelos de machine learning

```python
from skimage.transform import rotate
# Rotate the image 90 degrees clockwise
image_rotated = rotate(original, -90)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1]imshow(image_rotated)
ax[1].set_title("rotated")

fig.tight_layout()
plt.show()
```

```python
#re-escalando  una imagen
from skimage.transform import rescale
# Rescale the image to be 4 times smaller
image_rescaled = rescale(original, 1/4, anti_aliasing=True, multichannel=True)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(image_rescaled)
ax[1].set_title("rescaled")

fig.tight_layout()
plt.show()
```

```python
#cambiando el tamaño
from skimage.transform import resize
# Height and width to resize
height = 400
width = 500
# Resize image
image_resized = resize(original, (height, width), anti_aliasing=True)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(image_resized)
ax[1].set_title("resized")

fig.tight_layout()
plt.show()
```
